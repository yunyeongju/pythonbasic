```python
import seaborn as sns
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
def change_boolean(value):
    if value == 'yes':
        return True
    else:
        return False
titanic["alive"] = titanic["alive"].apply(change_boolean)
titanic.head()
```
